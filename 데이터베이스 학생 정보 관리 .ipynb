# 학생 정보 수정

```python
#Update
import sqlite3
def updata():
    con = sqlite3.connect("C:/Users/kowm6/Desktop/학생테이블.db")
    cur = con.cursor()

    # 수정할 학생의 학번 입력 받기
    HAKBUN = input("수정할 학생 학번 입력 : ")

    # 수정할 정보 입력 받기
    NAME = input("이름 : ")
    SEX = input("성별 : ")
    DEPT = input("전공 : ")
    HP = input("전화번호 : ")
    POST = input("우편번호 : ")
    ADDR = input("주소 : ")

    # 학생 테이블의 해당 학생 정보 수정
    sql = "UPDATE TSTUDENT SET NAME = '{}', SEX = '{}', DEPT = '{}', HP = '{}', POST = '{}', ADDR = '{}' WHERE HAKBUN = '{}';".format(NAME, SEX, DEPT, HP, POST, ADDR, HAKBUN)
    cur.execute(sql)

    con.commit()
    con.close()
```

# 학생 정보 삭제

```python
#DELETE

import sqlite3
def delete():
    con = sqlite3.connect("C:/Users/kowm6/Desktop/학생테이블.db")
    cur = con.cursor()

    # 삭제할 학생의 학번 입력 받기
    HAKBUN = input("삭제할 학생 학번 입력: ")

    # 입력 받은 학번을 가진 학생 정보 삭제하기
    sql = "DELETE FROM TSTUDENT WHERE HAKBUN = '{}';".format(HAKBUN)
    cur.execute(sql)

    
    con.commit()

    
    con.close()
```

# 학생 정보 검색

```python
#SEARCH

import sqlite3
def search():
    con = sqlite3.connect("C:/Users/kowm6/Desktop/학생테이블.db")
    cur = con.cursor()

   
    HAKBUN = input("검색할 학생 학번 입력: ")

    
    sql = "SELECT * FROM TSTUDENT WHERE HAKBUN = '{}';".format(HAKBUN)
    cur.execute(sql)

    # 검색 결과 출력하기
    result = cur.fetchone()
    if result:
        print("학번: ", result[0])
        print("이름: ", result[1])
        print("성별: ", result[2])
        print("전공: ", result[3])
        print("전화번호: ", result[4])
        print("우편번호: ", result[5])
        print("주소: ", result[6])
    else:
        print("해당하는 학생의 정보가 없습니다.")

    
    con.close()
```

# 학생 정보 목록 리스트

```python
#LIST
import sqlite3
def slist():
    con = sqlite3.connect("C:/Users/kowm6/Desktop/학생테이블.db")
    cur = con.cursor()

    sql = "select * from TSTUDENT;"

    cur.execute(sql)

    rows = cur.fetchall()

    for row in rows:
        print(row)

    con.close()
```
